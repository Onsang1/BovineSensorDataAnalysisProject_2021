## Introduction

To succeed in the dairy industry, dairy farms have to be very efficient because margins between cost and income are quite narrow. One key to success is to have longevous and healthy cows, with a high milk production level, that gets pregnant and start a new lactation every 13-15 months. There are several reasons why this can fail: cows can get sick, or have a lower production level than their pairs, or lower fertility and fail to get pregnant, and therefore fall under management strategies and decisions that make thea cow leave the herd.

Health disorders in dairy cows can happen at any time, but a substantial proportion of all health disorders occur in the early stage of the lactation (first 3-4 weeks postpartum). Becuase of this, the postpartum period (period after the cow gives birth to a baby cow, called calf) has became the key for cows and farmers since it can compromise herd health, welfare, and performance of the cows. Thus, events in this period may modify or increase the risk of culling and death and increase the cost due to necessary treatments. This will affect the cow's productive and reproductive performance in the future as well as the probability of staying in the herd as a productive cow.

In recent years, multiple devices have been developed and implemented by the dairy industry to automatically monitor behavioral, physiological, and performance parameters. Most of these devices have proven to be accurate and relayable while sensing and detecting changes in the parameters they mesure, and since many of these changes have been associated with helth disorders, they can alert when a cows is showing the warning patterns. However, although they can detect the changes in the parameters they are detecting, they can not confirm with a diagnostic and can only produce alert of changes. Noticeably, there is currently no commercially available software or device that can *predict* when the change in disease indicative parmeters will happen or what the health consecuences are going to be. 

Automated monitoring of cow behavior and physiological parameters using non-invasive sensors may help reduce the burden of health-monitoring programs. Sensor-generated data could be used alone or with traditional health-monitoring protocols to identify cows with health disorders. Furthermore, continuous monitoring of behavior, physiological, and performance parameters may allow for the detection of subtle changes before evident clinical signs appear. Earlier disease detection may benefit cows by preventing progression and improving response to treatment.

Research has shown before the impact of health disorders such as mastitis, metritis, displaced abomasum, ketosis, or pneumonia in the subsecuent performance of cows, and have also shown the relationship between the lower performance and the higher risk of culling, however, research that combines all this data from health disorders records and sensor data of behavior, phisiological, and performance parameters is hard to find and even more scarce research that uses it to predict the success and ultimately profitability of those animals.

Thus, we state the following **Research Questions:**

*1) Can cow features, sensor and non-sensor data of behavioral, physiological, and productivity parameters collected on the transition period of Dairy Cows (-15 to 30 relative to calving) determine the chances of those cows to succeed an get to the end of the lactation?*

*2) Can cow features, sensor and non-sensor data of behavioral, physiological, and productivity parameters collected on the transition period of Dairy Cows (-15 to 30 relative to calving) determine the productivity and/or profitability of cows that succeed to get to the end of their lactation?*


## Simple Understanding of a Cow’s Life:
Baby cow who had never been pregnant is called a **calf** (plural: calves). A female calf is called a **heifer**, while a male calf is called a **bull**.

If the calf is **male**, it will be sold early and exit the farm since it cannot produce milk for the purpose of dairy production.

If the calf is **female**, it will be fed with mother’s milk (colostrum) for 3 - 4 days to gain immunity against infection, then it will be transferred to a calf diet.

A heifer can start to be pregnant when it reaches about 2 years old. The total lifespan of a cow is about 12 years. She will be pregnant for about 9 months for each pregnancy. 

Before its first pregnancy, a heifer (female calf)  cannot produce milk. Once it gives birth to the first baby cow, the mother will enter the first **Days In Milk (DIM)**, where DIM = 0 on this day. It will be able to produce milk immediately after the birth of its baby. It will now officially be called a “**cow**”, and its **parity number = 1** (parity = how many times it has given birth to a baby cow).
Since its first birthing, a cow will be able to continuously produce milk. **However, the longer after its first birthing, the lower its milk production becomes.** Thus, to make the first calving to be the most sustainable and profitable, a cow should get pregnant between 2-4 months after calving. During its second pregnancy, milking will continue until it is **45 - 60 days** before its second parturition (birthing). This period is called the “**Dry Period**,” and the cows will not be milked, will be rested, and will be fed a special diet that prepares her body for the next lactation. Once it gives birth to its calf, it will start to be milked again. The **lactation** period then start with each calving and finish when the cow is dried.

In this study, we separated cows into:
- **parity = 1** (cows that only gave birth once)
- **parity = 2** (cows that only gave birth twice)
- **parity = 3+** (cows that gave birth 3 or more times), because the parameters of interest are not significantly different for cows beyond their third lactation and the number of individuals in each lactation group decreases dramatically after the third calving.

**Try PCA for the reduced model**

```python
# A function used to find the number of components for PCA that has the highest accuracy. It will return the optimal number of components for PCA and the maximum accuracy.
def pca_tester(test_ratio,model,predictors,outcome):
    num = 0
    max = 0
    
    # from 1 to 60 PCA components which one will have the highest accuracy score
    for i in range(1,60):
        reducer = PCA(n_components=i)
        log_predictors_red = reducer.fit_transform(predictors)

        PCA_mean = cross_val_score(model,log_predictors_red[:,:i],outcome,cv=test_ratio).mean()

        if(PCA_mean>max):
            max = PCA_mean
            num = i
            
    return [max,num]
```

```python
%%time

print("For a train-test split of 80/20, the best number of PCA components here is: "+str(pca_tester(10,lac_success_model_c1,log_predictors_c,y_outcome_c)[1]))
print("The mean accuracy for PCA-filtered Model in this case is: "+ str(round(pca_tester(10,lac_success_model_c1,log_predictors_c,y_outcome_c)[0],7)*100)+"%")
```

```python
%%time
print("For a train-test split of 60/40, the best number of PCA components here is: "+str(pca_tester(10,lac_success_model_c2,log_predictors_c,y_outcome_c)[1]))
print("The mean accuracy for PCA-filtered Model in this case is: "+ str(round(pca_tester(10,lac_success_model_c2,log_predictors_c,y_outcome_c)[0],7)*100)+"%")
```
